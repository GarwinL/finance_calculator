```python

import tkinter as tk
from tkinter import ttk
  
# parameters
HEIGHT = 700
WIDTH = 900 
LARGEFONT =("Verdana", 35)
  
class tkinterApp(tk.Tk):
     
    # __init__ function for class tkinterApp 
    def __init__(self, *args, **kwargs): 
         
        # __init__ function for class Tk
        tk.Tk.__init__(self, *args, **kwargs)
         
        # Creating Canvas 
        canvas = tk.Canvas(self, height=HEIGHT, width=WIDTH)
        canvas.pack()

        # Add menu bar
        menubar = tk.Menu(self)
        filemenu = tk.Menu(menubar, tearoff=0)
        #filemenu.add_command(label="Save", command=show_frame)
        #filemenu.add_command(label="Save as...", command=show_frame)
        #filemenu.add_command(label="Close", command=show_frame)
        filemenu.add_separator()
        filemenu.add_command(label="Exit", command=self.quit)
        menubar.add_cascade(label="File", menu=filemenu)

        tabmenu = tk.Menu(menubar, tearoff=0)
        tabmenu.add_command(label="Steuerrechner", command=lambda : self.show_frame(tax_calculator))
        tabmenu.add_command(label="Rentenrechner", command=lambda : self.show_frame(pension_calculator))
        #tabmenu.add_command(label="Dividendenrechner", command=show_frame)
        #tabmenu.add_command(label="bAV-Rechner", command=show_frame)
        menubar.add_cascade(label="Rechnerauswahl", menu=tabmenu)

        self.config(menu=menubar)
  
        # initializing frames to an empty array
        self.frames = {}
  
        # iterating through a tuple consisting
        # of the different page layouts
        for F in (tax_calculator, pension_calculator):
  
            frame = F(canvas, self)
  
            # initializing frames
            self.frames[F] = frame
  
        self.show_frame(tax_calculator)
  
    # to display the current frame passed as
    # parameter
    def show_frame(self, cont):
        frame = self.frames[cont]
        frame.tkraise()
```

```python
# Start page Steuerrechner
class tax_calculator(tk.Frame):
     
    def __init__(self, parent, controller):
         
        tk.Frame.__init__(self, parent)
        self.place(relx=0,rely=0, relwidth=1, relheight=1)
        label = tk.Label(self, text ="Steuerrechner", font=('Courier', 20))
        label.place(relx=0.3, rely=0, relwidth=0.4, relheight=0.05)
        
        # Entryfields --------------------------------------------------------------------
        #Parameter
        parameter_frame = tk.Frame(self, bg='#d9d9d9')
        parameter_frame.place(relx=0.05,rely=0.05, relwidth=0.2, relheight=0.9)

        nr_inputs = 8
        frame_gap = 0.01
        entry_height = (1.0-2*frame_gap)/(2*nr_inputs)
        cnt_inputs = 0

        # Jahreslohn
        label_sal = tk.Label(parameter_frame, text='Brutto-Jahreslohn', bg='#d9d9d9')
        label_sal.place(relx=frame_gap,rely=cnt_inputs*entry_height+frame_gap, relwidth=1.0-2*frame_gap, relheight=entry_height)
        cnt_inputs+=1
        entry_sal = tk.Entry(parameter_frame)
        entry_sal.place(relx=0.01,rely=cnt_inputs*entry_height+frame_gap, relwidth=1.0-2*frame_gap, relheight=entry_height)
        cnt_inputs+=1

        # Kirchensteuer
        label_chruchTax = tk.Label(parameter_frame, text='Kirchensteuer (BW/Bayern: 8%)', bg='#d9d9d9')
        label_chruchTax.place(relx=frame_gap,rely=cnt_inputs*entry_height+frame_gap, relwidth=1.0-2*frame_gap, relheight=entry_height)
        cnt_inputs+=1
        entry_chruchTax = tk.Entry(parameter_frame)
        entry_chruchTax.place(relx=0.01,rely=cnt_inputs*entry_height+frame_gap, relwidth=1.0-2*frame_gap, relheight=entry_height)
        cnt_inputs+=1

        # KV-Zusatzbeitrag
        label_kv = tk.Label(parameter_frame, text='KV-Zusatzbeitrag (in %)', bg='#d9d9d9')
        label_kv.place(relx=frame_gap,rely=cnt_inputs*entry_height+frame_gap, relwidth=1.0-2*frame_gap, relheight=entry_height)
        cnt_inputs+=1
        entry_kv = tk.Entry(parameter_frame)
        entry_kv.place(relx=0.01,rely=cnt_inputs*entry_height+frame_gap, relwidth=1.0-2*frame_gap, relheight=entry_height)
        cnt_inputs+=1

        # Steuerklasse
        label_retlost = tk.Label(parameter_frame, text='Steuerklasse', bg='#d9d9d9')
        label_retlost.place(relx=frame_gap,rely=cnt_inputs*entry_height+frame_gap, relwidth=1.0-2*frame_gap, relheight=entry_height)
        cnt_inputs+=1
        entry_retlost = tk.Entry(parameter_frame)
        entry_retlost.place(relx=0.01,rely=cnt_inputs*entry_height+frame_gap, relwidth=1.0-2*frame_gap, relheight=entry_height)
        cnt_inputs+=1

        result_frame = tk.Frame(self, bg='#d9d9d9')
        result_frame.place(relx=0.3,rely=0.15, relwidth=0.65, relheight=0.8)

        self.result_label = tk.Label(result_frame, bg='#d9d9d9', text = 'See Result here')
        self.result_label.place(relx=0.3,rely=0.4, relwidth=0.4, relheight=0.3)

        # Calculate Button
        button = tk.Button(self, text='Calculate', bg='#33cc33',font=40, command=lambda: self.show_result(entry_sal.get(), entry_chruchTax.get(), entry_kv.get()))
        button.place(relx=0.5, rely=0.05, relwidth=0.2, relheight=0.05)

    # Steuern: Basis 2021
    def tax_calculator(self, f_sal, f_annual=True):
        if(not f_annual):
            f_sal = f_sal*12.0

        if(f_sal<=9744):
            return 0
        elif(f_sal<=14753):
            y = (f_sal-9744)/10000
            return (995.21*y+1400)*y
        elif(f_sal<=57918):
            y = (f_sal-14753)/10000
            return (208.85*y+2397)*y+950.96
        elif(f_sal<=274612):
            return 0.42*f_sal-9136.63
        else:
            return 0.45*f_sal-17374.99

    # Arbeitslosenversicherung: 2,4% (50%)
    # Pflegeversicherung: 3,05% (50%) + Kinderloszuschlag (0,25%) - Beitragsbemessungsgrenze: 4837
    # Gesetzliche KV: 14,6% + Zusatzbeitrag (50%) - Beitragsbemessungsgrenze: 4837
    # Rentenversicherung: 18,6% (50%)
    def calc_social_tax(self, f_sal, f_kvAdd=1.3, f_annual=True):
        if(f_annual):
            l_care = min(f_sal,58044)*0.01775
            l_health = min(f_sal,58044)* 0.078
        else:
            l_care = min(f_sal,4837)*0.01775
            l_health = min(f_sal,4837)* (0.078 + f_kvAdd/2)

        l_jobless = f_sal*0.012
        l_pension = f_sal*0.093

        return l_jobless, l_care, l_health, l_pension

    # Solidaritätszuschlag
    def calc_soli(self, f_tax):
        l_soli = 0
        if(f_tax>16956): # Minderungszone
            l_soli = (f_tax-16956)*0.119
        elif(f_tax>31528):
            l_soli = f_tax*0.055
        return l_soli

    def show_result(self, f_sal, f_churchTax, f_kvAdd, f_annual=True):
        l_sal = float(f_sal)
        l_kvAdd = float(f_kvAdd)/100
        l_jobless, l_care, l_health, l_pension = self.calc_social_tax(l_sal, l_kvAdd, f_annual)
        l_socialTax = l_jobless + l_care + l_health + l_pension
        l_tax = self.tax_calculator(l_sal-0.92*l_socialTax) # 92% (2021, steigt um 2%/Jahr) der Sozialversicherung als Sonderausgaben zu berücksichtigen
        l_churchTax = float(f_churchTax)/100*l_tax # BW/Bayern: 8%, Rest: 9%
        l_soli = self.calc_soli(l_tax)
        l_overall = l_socialTax + l_tax + l_churchTax + l_soli

        self.result_label['text'] = 'Sozialabgaben: {}\nSteuern: {}\nKirchensteuer: {}\nSolidaritätszuschlag: {}\nGesamtabgaben: {}'.format(l_socialTax, l_tax, l_churchTax, l_soli, l_overall)
```

```python
# Start page Steuerrechner
class pension_calculator(tk.Frame):
     
    def __init__(self, parent, controller):
         
        tk.Frame.__init__(self, parent)
        self.place(relx=0,rely=0, relwidth=1, relheight=1)
        label = tk.Label(self, text ="Rentenrechner", font=('Courier', 20))
        label.place(relx=0.3, rely=0, relwidth=0.4, relheight=0.05)
        
        # Entryfields --------------------------------------------------------------------
        #Parameter
        parameter_frame = tk.Frame(self, bg='#d9d9d9')
        parameter_frame.place(relx=0.05,rely=0.05, relwidth=0.2, relheight=0.9)

        nr_inputs = 8
        frame_gap = 0.01
        entry_height = (1.0-2*frame_gap)/(2*nr_inputs)
        cnt_inputs = 0

        # Jahreslohn
        label_sal = tk.Label(parameter_frame, text='Brutto-Jahreslohn', bg='#d9d9d9')
        label_sal.place(relx=frame_gap,rely=cnt_inputs*entry_height+frame_gap, relwidth=1.0-2*frame_gap, relheight=entry_height)
        cnt_inputs+=1
        entry_sal = tk.Entry(parameter_frame)
        entry_sal.place(relx=0.01,rely=cnt_inputs*entry_height+frame_gap, relwidth=1.0-2*frame_gap, relheight=entry_height)
        cnt_inputs+=1

        # Arbeitsjahre
        label_workingyrs = tk.Label(parameter_frame, text='Arbeitsjahre', bg='#d9d9d9')
        label_workingyrs.place(relx=frame_gap,rely=cnt_inputs*entry_height+frame_gap, relwidth=1.0-2*frame_gap, relheight=entry_height)
        cnt_inputs+=1
        entry_workingyrs = tk.Entry(parameter_frame)
        entry_workingyrs.place(relx=0.01,rely=cnt_inputs*entry_height+frame_gap, relwidth=1.0-2*frame_gap, relheight=entry_height)
        cnt_inputs+=1

        # Rentenjahre
        label_retgrowth = tk.Label(parameter_frame, text='Rentenjahre', bg='#d9d9d9')
        label_retgrowth.place(relx=frame_gap,rely=cnt_inputs*entry_height+frame_gap, relwidth=1.0-2*frame_gap, relheight=entry_height)
        cnt_inputs+=1
        entry_retgrowth = tk.Entry(parameter_frame)
        entry_retgrowth.place(relx=0.01,rely=cnt_inputs*entry_height+frame_gap, relwidth=1.0-2*frame_gap, relheight=entry_height)
        cnt_inputs+=1

        # Vorzeitiger Renteneintritt
        label_retlost = tk.Label(parameter_frame, text='Vorzeitiger Renteneintritt(Jahre)', bg='#d9d9d9')
        label_retlost.place(relx=frame_gap,rely=cnt_inputs*entry_height+frame_gap, relwidth=1.0-2*frame_gap, relheight=entry_height)
        cnt_inputs+=1
        entry_retlost = tk.Entry(parameter_frame)
        entry_retlost.place(relx=0.01,rely=cnt_inputs*entry_height+frame_gap, relwidth=1.0-2*frame_gap, relheight=entry_height)
        cnt_inputs+=1

        # result frame
        result_frame = tk.Frame(self, bg='#d9d9d9')
        result_frame.place(relx=0.3,rely=0.15, relwidth=0.65, relheight=0.8)

        self.result_label = tk.Label(result_frame, bg='#d9d9d9', text = 'See Result here')
        self.result_label.place(relx=0.3,rely=0.4, relwidth=0.4, relheight=0.3)

        # Calculate Button
        button = tk.Button(self, text='Calculate', bg='#33cc33',font=40, command=lambda: self.calculator(entry_workingyrs.get(), entry_sal.get()))
        button.place(relx=0.5, rely=0.05, relwidth=0.2, relheight=0.05)

    # Calculation of monthly rent: 
    # Monatliche Rentenhöhe = Entgeltpunkte x Zugangsfaktor x aktueller Rentenwert x Rentenartfaktor
    # Entgeltpunkte: Rentenpunkte
    # aktueller Rentenwert: 2020: 1 Punkt = 34.19€ Rente/Monat
    # Rentenartfaktor: Alter/volle Erbwerbsminderung = 1.0, teilweise Erbwerbsminderungsrente = 0.5, Halbwaisenrente = 0.1
    # Zugangsfaktor: Zu- (später Renteneintritt) und Abschläge (früherer Renteneintritt: 0.3%/Monat), ohne = 1.0
    def calculator(self, f_working_yrs, f_input_sal, f_average_sal=40551.0, f_pensionpoints_value=34.19):
        l_working_yrs = float(f_working_yrs)
        l_input_sal = float(f_input_sal)
        rent_val = (l_input_sal/f_average_sal)*l_working_yrs*f_pensionpoints_value
        self.result_label['text'] = str(rent_val)
```

```python
# Driver Code
app = tkinterApp()
app.mainloop()
```
